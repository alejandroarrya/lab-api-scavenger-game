```python
#Obtaining a list of all the forks requires a new request. This new request is transformed into 'json' and, later
#into a dataframe.

forks = requests.get('https://api.github.com/repos/ironhack-datalabs/mad-oct-2018/forks')
forks = forks.json()
forks_df = pd.DataFrame(forks)
forks_df.head()
```

```python
#We don't need to do a loop to find out about the different languages used in all forks. A set will suffice.
#We can turn the set into an array(list) afterwards.

languages = list(set(forks_df['language']))
print(languages)
```
